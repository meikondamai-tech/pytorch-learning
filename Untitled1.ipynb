```python
# 28x28のダミー画像データを作成 (0.0〜1.0の範囲)
# 実際のMNISTデータは0〜255ですが、学習では0〜1に正規化するのが一般的です
image_tensor = torch.rand(28, 28)

# 作成したtensorの形状と一部の中身を確認
print("Tensorの形状 (Shape):", image_tensor.shape)
print("-" * 20)
print("Tensorの中身（左上の一部）:")
print(image_tensor[:3, :5]) # 最初の3行と5列だけ表示
```

```python
# 形状を (28, 28) から (1, 28, 28) へ
image_with_channel = image_tensor.unsqueeze(0)

# 形状を (1, 28, 28) から (1, 1, 28, 28) へ
image_final = image_with_channel.unsqueeze(0)

print("チャンネル次元追加後の形状:", image_with_channel.shape)
print("最終的なネットワーク入力形状:", image_final.shape)
```
